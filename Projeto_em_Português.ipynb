```python
# Constantes
NOTA_MINIMA = 0.0
NOTA_MAXIMA = 10.0
NOTA_APROVACAO = 7.0

# Função para adicionar nota
def adicionar_nota():
    """Adiciona uma nova nota à lista de notas do aluno."""
    notas = []
    while True:
        try:
            nota = float(input(f"Digite a nota do aluno (entre {NOTA_MINIMA} e {NOTA_MAXIMA}, ou -1 para finalizar): "))
            if nota == -1:
                if len(notas) == 0:
                    print("Você deve adicionar pelo menos uma nota.")
                    continue
                break
            if NOTA_MINIMA <= nota <= NOTA_MAXIMA:
                notas.append(nota)
            else:
                print(f"Por favor, insira uma nota entre {NOTA_MINIMA} e {NOTA_MAXIMA}.")
        except ValueError:
            print("Entrada inválida. Digite um número.")
    return notas

# Função para calcular média
def calcular_media(notas):
    """Calcula a média das notas do aluno."""
    return sum(notas) / len(notas) if notas else 0

# Função para verificar situação
def verificar_situacao(media):
    """Verifica a situação do aluno (aprovado ou reprovado) com base na média."""
    return "Aprovado" if media >= NOTA_APROVACAO else "Reprovado"

# Função para exibir o relatório do aluno
def exibir_relatorio(nome, notas, media, situacao):
    """Exibe um relatório completo das notas do aluno."""
    print("\nRelatório do Aluno:")
    print(f"Nome: {nome}")
    print(f"Notas: {notas}")
    print(f"Média: {media:.2f}")
    print(f"Situação: {situacao}")
    print(f"Maior Nota: {max(notas):.2f}")
    print(f"Menor Nota: {min(notas):.2f}")
    print(f"Notas Acima da Média: {sum(1 for n in notas if n > media)}")

# Função para salvar os dados dos alunos em um arquivo
def salvar_dados(nome, notas, media, situacao):
    """Salva os dados do aluno em um arquivo de texto."""
    with open("dados_alunos.txt", "a") as arquivo:
        arquivo.write(f"Nome: {nome}\n")
        arquivo.write(f"Notas: {notas}\n")
        arquivo.write(f"Média: {media:.2f}\n")
        arquivo.write(f"Situação: {situacao}\n\n")

# Função principal para gerenciar o sistema de notas
def sistema_notas():
    """Sistema principal para gestão de notas de vários alunos."""
    alunos = {}

    while True:
        print("\nMenu:")
        print("1. Adicionar notas para novo aluno")
        print("2. Exibir relatório de um aluno")
        print("3. Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            nome_aluno = input("Digite o nome do aluno: ").strip()
            if nome_aluno in alunos:
                print("Este aluno já foi adicionado.")
                continue

            notas = adicionar_nota()
            media = calcular_media(notas)
            situacao = verificar_situacao(media)

            alunos[nome_aluno] = {'notas': notas, 'media': media, 'situacao': situacao}
            salvar_dados(nome_aluno, notas, media, situacao)
            print(f"\nNotas do aluno {nome_aluno} foram salvas com sucesso.")

        elif escolha == "2":
            nome_aluno = input("Digite o nome do aluno para exibir o relatório: ").strip()
            if nome_aluno in alunos:
                dados_aluno = alunos[nome_aluno]
                exibir_relatorio(nome_aluno, dados_aluno['notas'], dados_aluno['media'], dados_aluno['situacao'])
            else:
                print(f"Aluno '{nome_aluno}' não encontrado.")

        elif escolha == "3":
            print("Saindo do sistema.")
            break

        else:
            print("Opção inválida. Tente novamente.")

# Executar o sistema
if __name__ == "__main__":
    sistema_notas()
```

O sistema de gerenciamento de notas dos alunos opera em várias etapas, organizadas em funções:

 Adicionar notas: A função add_grade() exige que o usuário insira as notas do aluno. As notas são adicionadas a uma lista até que o usuário digite -1, com validação para garantir que as notas estejam no intervalo correto (0 a 10).
 Calcular média: A função calculate_average() calcula a média das notas inseridas, somando todas elas e dividindo pelo número de notas.
 Verificar status: A função check_status() basicamente examina se o aluno foi aprovado ou reprovado. Se a média for maior ou igual a 7, o aluno é "Aprovado", caso contrário, é "Reprovado".
 Exibir relatório: A função display_report() mostra as informações resumidas do aluno, como nome, notas, média, status, nota mais alta e mais baixa e quantas notas estão acima da média.
 Salvar dados: A função save_student_data() salva os dados do aluno em um arquivo de texto para persistência.
 Gerenciamento por meio de um menu: A função primária grade_system() permite que o usuário adicione alunos, visualize relatórios e saia do sistema usando um menu interativo.

Isso significa que é fácil gerenciar as notas de diferentes alunos e que o sistema não é carregado, proporcionando uma estrutura clara e eficiente.
